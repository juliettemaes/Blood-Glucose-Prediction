### Data pre-preparation

#### In time series data, timestamp information is very important to capture features such as periodicity, trend, etc. of time. We transform the time column and extract ‘hour’ as a new feature to add to the data

```python
# date conversion
train.time = pd.to_datetime(train.time, format='%H:%M:%S')
test.time = pd.to_datetime(test.time, format='%H:%M:%S')
train['hour'] = train['time'].dt.hour
test['hour'] = test['time'].dt.hour
```

##### Processing of missing values for 'bg'
Le remplissage de la bg n'est pas mesuré toutes les 10 minutes

```python
train = pd.read_csv('./data/train.csv',header=0)
test = pd.read_csv('./data/test.csv',header=0)
```

```python
import pandas as pd

def fill_na_with_avg(df, prefixe):

    cols = [col for col in df.columns if col.startswith(prefixe)]
    

    for col in cols:

        if pd.isna(df[cols[0]]).any() and pd.isna(df[cols[1]]).any():
            df[cols[0]] = df[cols[0]].fillna(df[cols[2]])  
            df[cols[1]] = df[cols[1]].fillna(df[cols[2]])  
            

            for i in range(2, len(cols) - 3, 3):
                prev_col = df[cols[i]]
                next_col = df[cols[i + 3]]
                df[cols[i + 1]] = df[cols[i + 1]].fillna((prev_col + next_col) / 2)
                df[cols[i + 2]] = df[cols[i + 2]].fillna((prev_col + next_col) / 2)

    return df
```

```python
import pandas as pd

def fill_na_with_avg2(df, prefixe):

    cols = [col for col in df.columns if col.startswith(prefixe)]
    
    for col in cols:        
        if pd.isna(df[cols[0]]).any() :
            df[cols[0]] = df[cols[0]].fillna(df[cols[1]])  
            

            for i in range(1, len(cols) - 3, 3):
                prev_col = df[cols[i]]
                next_col = df[cols[i + 3]]
                
                df[cols[i + 1]] = df[cols[i + 1]].fillna((prev_col + next_col) / 2)
                df[cols[i + 2]] = df[cols[i + 2]].fillna((prev_col + next_col) / 2)
    return df
```

```python
train = fill_na_with_avg(train, 'bg')
train = fill_na_with_avg2(train, 'bg')
```

```python
import pandas as pd

def fill_na_with_linear_interpolation(df, prefixe):
    # 筛选以 prefixe 开头的所有列
    cols = [col for col in df.columns if col.startswith(prefixe)]
    
    # 对所有的 'bg' 列进行线性插值
    df[cols] = df[cols].interpolate(method='linear', axis=0, limit_direction='both')
    
    return df
```

```python
train = fill_na_with_linear_interpolation(train, 'bg')
```

```python
bg_cols = [col for col in train.columns if 'bg' in col]
nan_summary = train[bg_cols].isnull().sum()
print(nan_summary)
```

##### Processing of missing values for 'insulin'

```python
def fill_na_with_neighbors(df, prefixe):
    cols = [col for col in df.columns if col.startswith(prefixe)]
    df[cols] = df[cols].apply(lambda col: col.interpolate(method = 'linear', limit_direct = 'both'))

    return df
```

```python
train = fill_na_with_neighbors(train, 'insulin')
```

```python
insulin_cols = [col for col in train.columns if 'insulin' in col]
nan_summary = train[insulin_cols].isnull().sum()
print(nan_summary)
```
